```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd
```

```python
def f(x, y):
    return np.exp(x) * (1 - y)

y0 = 2
```

```python
a = 0
b = 1
h = 0.01

n = int((b - a) / h) + 1

x = np.linspace(a, b, n)
y = np.zeros_like(x)
y[0] = y0

## Implementing the Euler method algorithm
for i in range(1, n):
    y[i] = y[i-1] + h * f(x[i-1], y[i-1])
c = (y0 - 1) / np.exp(0)
ya = 1 + c * np.exp(-x)
```

Plugging in the retrieved data into a table and plotting onto a graph

```python
print(' x    | Numerical | Analytic ')
print('------+-----------+----------')
for i in range(n):
    print('{:.2f} |  {:.6f} | {:.6f}'.format(x[i], y[i], ya[i]))
plt.plot(x, ya, label='Analytic')
plt.plot(x, y, label='Numerical')
plt.legend()
plt.show()
```

```python
def rk4(x,y):
  k1 = f(x,y)
  k2 = f(x+(1/2)*h,k1*y)
  k3 = f(x+(1/2)*h,k2*y)
  k4= f(x+(1/2)*h,k3*y)
  y1 = y + (k1 + 2*k2 +2*k3 + k4)*h/6
  return y1
xr = np.linspace(a,b,n)
yr = np.zeros_like(xr)
xactual = np.linspace(0,101,1)
def actual(x):
  ya = 1 + c * np.exp(-x)
  return ya
plt.plot(f(xr,yr))
plt.plot(xactual,actual(xactual),'m')
```

Above we have the Runge-Kutta 4th order method. This method uses iterative step values to approximate the solution to the ordinary differential equation.
